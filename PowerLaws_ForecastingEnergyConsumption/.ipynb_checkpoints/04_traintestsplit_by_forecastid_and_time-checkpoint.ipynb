## Create Train-Test Split

```python
def my_train_split(df, train_size=2/3):
    split_point = int(df.shape[0] * train_size)
    train = df.iloc[:split_point,:]
    return train

def my_test_split(df, train_size=2/3):
    split_point = int(df.shape[0] * train_size)
    test = df.iloc[split_point:,:]
    return test
```

```python
df_gb = df_15.groupby(['ForecastId'], as_index=False)
```

```python
train = df_gb.apply(my_train_split)
```

```python
test = df_gb.apply(my_test_split)
```
