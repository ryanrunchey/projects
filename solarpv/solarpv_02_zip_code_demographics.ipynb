# Import Libraries

```python
import pandas as pd
```

# Zip Code Demographic Data

```python
# import CDXTECH zip code demographic data

filepath = "CDXTECH/CDETECH_ZIP.csv"
df_zip = pd.read_csv(filepath)

df_zip.head(3)
```

```python
# drop first row of extraneous NaN values
# rename first column 'Zip Code'

df_zip.drop(labels=0, axis=0, inplace=True)
df_zip.rename(columns={'Your Entry': 'Zip Code'}, inplace=True)
df_zip.head(3)
```

```python
# inspect DataFrame info

df_zip.info()
```

```python
# clean up to keep desired columns

cols = ['Zip Code']
y = [x for x in df_zip.columns if type(df_zip[x][1]) != str]
cols.extend(y)

rmv = list(filter(lambda col: (('FIPS' in col) or ('CBSA' in col) or ('Growth' in col) or ('Growing' in col)), cols))
cols = [x for x in cols if x not in rmv]        

df_zip = df_zip[cols]


# dropna values
df_zip = df_zip.dropna()


# review info
df_zip.info()
```

```python
# set index to 'Zip Code'

df_zip.set_index('Zip Code', inplace=True)
df_zip.head(3)
```

```python
# pickle cleaned-up df_zip

df_zip.to_pickle('02_df_zip.pkl')
```
