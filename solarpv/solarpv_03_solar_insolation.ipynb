# Import Libraries

```python
import pandas as pd
```

# Google Project Sunroof

```python
filepath = "google_projectsunroof/solar_potential_by_postal_code.csv"

df_solar = pd.read_csv(filepath)
df_solar.head(3)
```

```python
df_solar.info()
```

```python
df_solar = df_solar.dropna()
df_solar.shape
```

```python
# cast region_name astype int because this is Zip Code
# set_index to region_name

df_solar['region_name'] = df_solar['region_name'].astype(int, copy=False)
df_solar['region_name'] = df_solar['region_name'].astype(str, copy=False)
df_solar = df_solar.set_index('region_name')
df_solar.head(2)
```

```python
# select state_name == California

df_solar_ca = df_solar[df_solar['state_name'] == 'California']
df_solar_ca.shape
```

```python
# drop extraneous columns

col_drop = ['state_name','lat_max', 'lat_min', 'lng_max', 'lng_min', 'install_size_kw_buckets',\
            'carbon_offset_metric_tons', 'existing_installs_count']

df_solar_ca = df_solar_ca.drop(col_drop, axis=1)
df_solar_ca.shape
```

```python
# pickle DataFrame

df_solar_ca.to_pickle('03_df_solar_ca.pkl')
```
