## Merge DataFrames

```python
# merge df_capacity with df_zip

df_merge_1 = df_capacity.merge(df_zip, how='left', left_index=True, right_index=True)
df_merge_1.head(3)
```

```python
df_merge_1.shape
```

```python
df_merge_1.info()
```

```python
# merge df_merge_1 with df_solar_ca

df_merge_2 = df_merge_1.merge(df_solar_ca, how='left', left_index=True, right_index=True)
df_merge_2.head(3)
```

```python
df_merge_2.shape
```

```python
df_merge_2.info()
```

```python
# df_zip does not contain information on these Zip Codes

df_merge_2[df_merge_2['Population'].isnull()]
```

```python
# Google Project SunRoof has data on 928 California Zip Codes representing 89.1% of installed capacity
# We can drop the 10.9% of data that is not represented by Google Project Sunroof
# OR we can use a tree-based model(s) which handle NaN values / but then we can't use OLS baseline

df_merge_2[df_merge_2['kw_total'].isnull()]['System Size AC'].sum() / df_merge_2['System Size AC'].sum()
```

```python
# drop NaN

df_ml = df_merge_2[~df_merge_2.isnull().any(axis=1)]
df_ml.shape
```

```python
# validate all NaN dropped

df_ml[df_ml.isnull().any(axis=1)]
```
